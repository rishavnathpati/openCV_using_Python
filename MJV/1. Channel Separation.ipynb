```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
path=r'D:/Study/Python/openCV/'
```

```python
img=cv2.imread(path+'bgr.png')
img.shape
img_color=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
plt.title("Colour")
plt.imshow(img_color)

plt.subplot(2,2,2)
plt.title("Red")
plt.imshow(img_color[:,:,0],'gray')

plt.subplot(2,2,3)
plt.title("Green")
plt.imshow(img_color[:,:,1],'gray')

plt.subplot(2,2,4)
plt.title("Blue")
plt.imshow(img_color[:,:,2],'gray')
```
