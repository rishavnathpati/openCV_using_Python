GAUSSIAN BLUR AND SOBEL

```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
import math
path=r'D:/Study/Python/openCV/College/res/'
```

```python
def show_img(img):
    fig=plt.figure(figsize=(7,7))
    ax=fig.add_subplot(111)
    plt.imshow(img,'gray')
```

```python
img=cv2.imread(path+'lena.tif',0)
height,width=img.shape
border_img=cv2.copyMakeBorder(img,1,1,1,1,cv2.BORDER_CONSTANT,value=0)
```

```python
blur_kernal=np.array([[1,2,1],[2,4,2],[1,2,1]])
sobel_x=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobel_y=np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

blur_img=np.zeros(img.shape,'uint8')
sobel_edge_img=np.zeros(img.shape,'uint8')
```

```python
for i in range(height):
    for j in range(width):
        blur_img[i,j]=int(np.sum(border_img[i:i+3,j:j+3]*blur_kernal)/np.sum(blur_kernal))
```

```python
show_img(blur_img)
sobel_border_img=cv2.copyMakeBorder(blur_img,1,1,1,1,cv2.BORDER_CONSTANT,value=0)
```

```python
for i in range(height):
    for j in range(width):
        sum_x=np.sum(sobel_border_img[i:i+3,j:j+3]*sobel_x)
        sum_y=np.sum(sobel_border_img[i:i+3,j:j+3]*sobel_y)
        sobel_edge_img[i,j]= round(math.sqrt(sum_x*sum_x+sum_y*sum_y))
```

```python
show_img(sobel_edge_img)
```
