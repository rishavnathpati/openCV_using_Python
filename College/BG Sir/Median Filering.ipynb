```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
import math
path=r'D:/Study/Python/openCV/College/res/'
```

```python
def show_img(img):
    fig=plt.figure(figsize=(10,10))
    ax=fig.add_subplot(111)
    plt.imshow(img,'gray')
```

```python
img=cv2.imread(path+'img3.png',0)
show_img(img)
height,width=img.shape
border_img=cv2.copyMakeBorder(img,2,2,2,2,cv2.BORDER_CONSTANT,value=0)

blur_img=np.zeros(img.shape,'uint8')
median_arr=np.zeros((5,5))
```

```python
for i in range(height):
    for j in range(width):
        median_arr=border_img[i:i+5,j:j+5]
        np.reshape(median_arr,25)
        np.sort(median_arr)
        blur_img[i,j]=np.median(median_arr)
```

```python
show_img(blur_img)
```
