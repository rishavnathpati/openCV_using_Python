```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random

path=r'D:/Study/Python/openCV/College/res/'
```

```python
img=cv2.imread(path+"256col.png")
img=cv2.resize(img,(200,200),0,0,cv2.INTER_NEAREST)
inf=255**3
k=5
a=[]
for i in range(3):
    a.append(random.sample(range(256),k))
a=np.transpose(a)
a_=np.copy(a)
x,y,z=img.shape
b=np.ones((x,y),'uint8')*-1
out=np.zeros(img.shape,'uint8')
im1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(im1)
```

```python
f=0
b_=np.ones(b.shape)
while(np.sum(np.absolute(b-b_))>5):
    b_=np.copy(b)
    f+=1
    c=np.zeros((k,3))
    d=np.zeros(k)
    for i in range(x):
        for j in range(y):
            m=inf
            for l in range(k):
                m1=np.sum(np.square(img[i][j]-a[l]))
                if m>m1:
                    m=m1
                    ind=l
            b[i][j]=ind
            d[ind]+=1
            c[ind]+=img[i][j]
    a=c/d[:,None]
```

```python
for i in range(x):
    for j in range(y):
        out[i][j]=a_[b[i][j]]
out=cv2.cvtColor(out,cv2.COLOR_BGR2RGB)
plt.imshow(out)
```
