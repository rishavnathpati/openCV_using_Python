```python
img=cv2.imread(path+'crossword.jpg',0)
plt.imshow(img,'gray')
```

```python
def show_pic(img):
    fig=plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,'gray')
```

```python
show_pic(img)
```

```python
ret, thresh= cv2.threshold(img,150,255,cv2.THRESH_BINARY)
show_pic(thresh)
```

```python
thresh2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,5,8)
show_pic(thresh2)
```
