```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
blank_img=np.zeros(shape=(512,512,3),dtype=np.int16)
blank_img.shape
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,220,0),thickness=3)
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(233,0,0),thickness=5)
#cv2.line(blank_img,pt1=(512,0),pt2=(0,512),color=(233,0,0),thickness=5)
cv2.circle(blank_img,center=(400,100),radius=50,color=(0,0,100),thickness=-1)
cv2.circle(blank_img,center=(100,400),radius=70,color=(0,0,100),thickness=3)
cv2.putText(blank_img,text='Hello openCV',org=(10,500),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=2,color=(0,100,0),thickness=4,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
vertices
```

```python
blank_img=np.zeros(shape=(512,512,3),dtype=np.int32)

# Draw polygon
vertices=np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
#vertices
#vertices.shape
pts=vertices.reshape((-1,1,2))
pts.shape
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,255),thickness=5)
plt.imshow(blank_img)
```
