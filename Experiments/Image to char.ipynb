```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
import math
path=r'D:/Study/Python/openCV/College/res/'
```

```python
def show_img(img):
    fig=plt.figure(figsize=(10,10))
    ax=fig.add_subplot(111)
    plt.imshow(img,'gray')
```

```python
img=cv2.imread(path+'lena.tif',0)
```

```python
edge=cv2.Canny(img,100,100)
show_img(edge)
```

```python
resized=cv2.resize(edge,(64,64))
show_img(resized)
```

```python
ret,thresh1 = cv2.threshold(resized,10,255,cv2.THRESH_BINARY)
show_img(thresh1)
```

```python
charimg=np.zeros((64,64),'U1')
charimg.fill('@')
charimg
for i in range(thresh1.shape[0]):
    for j in range(thresh1.shape[1]):
        if thresh1[i][j]==0:
            charimg[i][j]='.'
```
